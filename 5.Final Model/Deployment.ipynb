```python
import pickle
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import f_classif, SelectFdr, SelectKBest, SelectPercentile, chi2
```

```python
dataset=pd.read_csv("Heart_Disease_Prediction.csv")
ds=pd.get_dummies(dataset, drop_first=True, dtype=int)
```

```python
Indep=ds.drop('Heart Disease_Presence', axis=1)
dep=ds['Heart Disease_Presence']
```

```python
#Feature Selection k_best
best = SelectKBest(score_func=chi2, k=5)
fit1=best.fit(Indep,dep)
Indep=fit1.transform(Indep)

x_train,x_test,y_train,y_test= train_test_split(Indep, dep, test_size=0.30, random_state=0)
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
print(x_train.shape)
print(x_test.shape)
```

```python
loaded_model=pickle.load(open("Final_model_KNN.sav",'rb'))
loaded_model
```
